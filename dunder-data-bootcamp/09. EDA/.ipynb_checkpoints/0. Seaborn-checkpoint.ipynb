This notebook will dive into the Seaborn visualization library in Python. This notebook assumes no previous visualization knowledge.

The [Python Seaborn library](http://seaborn.pydata.org/index.html) is a high-level, easy-to-use interface for creating powerful and beautiful visualizations. Seaborn does not actually perform any of the visualizations but instead, calls the primary Python visualization library, matplotlib, to do all the heavy lifting. Sometimes, Seaborn is referred to as a **wrapper** for matplotlib, which is a more difficult library to use but provides more fine-grained control.

The seaborn documentation is excellent and you well be well-served to read all of it. The library is fairly minimal and only exposes a relatively few amount of functions. I suggest you start with plotting sections of the [seaborn tutorial](http://seaborn.pydata.org/tutorial.html#). The tutorial separates out the plotting functions into three separate sections - distributions, linear relationships and categorical data.

## Seaborn integration with pandas
All seaborn plotting functions contain a **`data`** parameter that accepts a pandas DataFrame. This allows you to use the string names of the DataFrames for the other arguments in the plotting functions.

## Seaborn and Tidy data
Seaborn plotting functions work best with tidy data. Tidy data makes plotting with seaborn incredibly easy.

## Axes vs Grid seaborn plots
All of seaborn plotting functions return either an 'axes' or a 'grid'. 

**axes** are simpler plots that contain a single panel (a single x and y axis). They are technically matplotlib axes objects. Most seaborn plotting functions return axes.

**grids** are more complex plots that contain multiple panels (many different plots) and are technically seaborn objects.

## Check the seaborn API
Many Python libraries will have a link for the **API** (application programming interface). This will be a page that lists all the top-level functions and classes. Open up the the [seaborn API](http://seaborn.pydata.org/api.html) to see all the plotting functions. Click on several of the plotting functions in the **distribution, regression, categorical, matrix, and timeseries** plots. Determine whether an axes or a grid is returned. ## Grid functions create multiple plots by subsetting data
Keep the [Seaborn API]((http://seaborn.pydata.org/api.html) open for the rest of the notebook.

The plotting functions that produce grids are at a higher level than the axes functions. These grid functions can produce all the other plots in their respective API section. They have the ability to create several plots at once by  subsetting on other variables in the dataset. For instance, **`factorplot`** is like a grid plotting function that uses the rest of the categorical plots (point, bar, count, box, violin, strip, swarm).

## Getting started with Axes vs Grid plotting functions
An example should clear up the differences between grid and axes functions. Below we will first create a plot using the categorical axes plotting functions (point, bar, count, box, violin, strip, swarm) and then again with the grid function **`factorplot`**.

## Using Seaborn data
Conveniently, seaborn supplies the **`load_dataset`** function to directly load a pandas DataFrame. Use the function **`get_dataset_names`** to get a list all the datasets seaborn has available.

We will use the **`tips`** dataset to begin.

```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

### boxplot of tips by sex
**`boxplot`** is a categorical plot that creates the a single panel axes plot. The below compares tips between males and females. Pass the column names as strings to the **`x`** and **`y`** parameters and the pandas DataFrame to the **`data`** parameter.

```python
sns.boxplot(x='sex', y='tip', data=tips)
```

## Use `factorplot` to create a grid of box plots
**`factorplot`** creates a grid of categorical plots. Here we will create a box plot of tips for males and females for each level of smoker (Yes/No). We can use the **`row`** or **`column`** parameter to subset the data further. Use the **`kind`** parameter to select the type of categorical plot.

```python
sns.factorplot(x='sex', y='tip', row='smoker', data=tips, kind='box')
```

## Add more dimensionality with hue
The **`hue`** parameter is available to add another layer of dimensionality within the same panel. Here we add the time of the meal (lunch/dinner).

```python
sns.boxplot(x='sex', y='tip', hue='time', data=tips)
```

## 5 layers of dimensionality with `factorplot`
Along with **`hue`**, seaborn allows grid plotting functions to use **`row`** and **`col`** parameters which along with **`x`** and **`y`** total 5 dimensions for very detailed visual slicing and comparing of data.

```python
# this dataset isn't quite dense enough to have data for every combination.
# Notice how there are no male, non-smokers on friday at lunch
sns.factorplot(x='sex', y='tip', hue='time', row='smoker', col='day', data=tips, kind='box')
```

```python
# check that there are infact no male, non-smokers at lunchtime on friday
tips.query('sex=="Male" and smoker=="No" and time =="Lunch" and day=="Fri"')
```

## Compare axes vs grid plots with violinplot
We will do the same comparison between axes and grid plots using violin plots

```python
# switch categorical from x to y axis - this makes the plot horizontal
sns.violinplot(x='tip', y='sex', hue='day', data=tips)
```

```python
# create a grid, subset by time
sns.factorplot(x='tip', y='sex', hue='day', col='time', data=tips, kind='violin')
```

# Categorical vs Regression Plots
Generally speaking, there are two broad classifications of variables (the columns) - **categorical** and **continuous**. Continuous variables are those that can take on an infinite number of possible values. Categorical variables are those that can only take on a finite number of values.

Numeric variables can be either continuous or categorical (like movie rating or stage of cancer). Each discrete value of a categorical variable is sometimes referred to as a **level**. Categorical variables are sometimes referred to as factors or discrete variables. Continuous variables are also referred to as regression variables.

Categorical plots typically have all the possible categories on the x axis with a corresponding continuous value on the y-axis.

Regression plots on the other hand have continuous values on both the x and y axis.

## Regression plots: axes vs grid plots
The differences between axes and grid plots should now be clear. axes are simpler and create a single plot. grids allow for creation of multiple panels by subsetting with other variables.

We can see the exact same thing with regression plots. Regression meaning that both x and y are numerical (usually continuous) variables. **`regplot`** creates an axes while **`lmplot`** creates an grid.

```python
tips.head()
```

```python
# one panel. returns axes. no hue parameter. simple
sns.regplot(x='total_bill', y='tip', data=tips)
```

```python
# multiple panels by time, plus has hue parameter.
sns.lmplot(x='total_bill', y='tip', hue='sex', col='time', data=tips)
```

## Distribution plots: axes vs grid
Just as in the categorical and regression plots section of the API, the distribution plots have both axes and grid functions. The distribution plots are unique in that they have two grid plotting functions, **`joinplot`** and **`pairplot`**. **`joinplot`** shows the distributions of two variables while **`pairplot`** shows the joint distribution of every variable pair possible.

We will first make some axes plots with **`distplot`**, **`kdeplot`** and **`rugplot`** and then do the equivalent with the two grid plots.

```python
tips.head()
```

```python
# create a histogram. By default it adds the kde (kernel density estimate)
sns.distplot(tips.tip)
```

```python
# kde alone
sns.kdeplot(tips.tip)
```

```python
# kde can do bivariate distributions
sns.kdeplot(tips.tip, tips.total_bill)
```

```python
# rugplot
sns.rugplot(tips.tip)
```

### `distplot` can do it all
**`distplot`** has boolean parameters **`rug`**, **`hist`** and **`kde`** to do all three plots at once.

```python
sns.distplot(tips.tip, rug=True, hist=True, kde=True)
```

## Distribution grid plots
**`joinplot`** plots two variables against one another and adds some kind of distribution plot outside of both the x/y axis. It outputs a grid.

```python
sns.jointplot('tip', 'total_bill', data=tips)
```

**`pairplot`** creates a plot for every regression (numeric) variable combination. The diagonal plot is a univariate summary of each variable.

```python
sns.pairplot(tips, hue='smoker')
```

## Learning just enough matplotlib
As was mentioned previously, seaborn is an interface to matplotlib and essentially figures out all the matplotlib for you so you don't have to bang your head against the wall for hours figuring out the minute details. Take a look at the basic anatomy of a matplotlib figure. There are two main objects, the **`figure`** and the **`axes`**. An **`axes`** is simply a single plot, not to be confused with the x or y axis. The figure can contain multiple axes. 

![matplotlib anatomy](images/fig_map.png)

The basic seaborn plotting functions, the ones in this tutorial called **`axes`** are actually returning a matplotlib **`axes`** object.

To modify the matplotlib axes object from a seaborn axes plot, we will save the result to a variable, **`ax`**. This is a matplotlib convention.

After saving the result to **`ax`**, we will use the dot notation to access the different parts of the plot, such as the x or y axis, the x or y axis labels, the title, the legend and so forth. Most of these methods controlling these parts of the plots begin with **`set_`**.

We can also modify the matplotlib figure. We can access the figure through **`ax.figure`** which has its own **`set_`** methods to modify itself.

The **`grid`** seaborn plots on the other hand have parameters **`size`** and **`aspect`** that control the size of each panel. These functions technically regurn a seaborn grid object which contains matplotlib figures and axes objects as well and can also be access through the dot-notation.

```python
# default size (6,4)
sns.swarmplot(x='sex', y='tip', data=tips)
```

### Modify figure with matplotlib `set_` methods
We will first save the result of the swarm plot to **`ax`**, which is a matplotlib axes object. We will then use **`set_`** methods to modify the plot. Notice that to set the size of the entire figure, you must access it with **`ax.figure`**.

```python
# save result to ax. ax is
ax = sns.swarmplot(x='sex', y='tip', data=tips)

# change figure size, x and y labels and create a title
ax.figure.set_size_inches(12, 8)
ax.set_ylabel('Tip Amount')
ax.set_xlabel('Gender')
ax.set_title('Tip Amount by Gender')

# grab onto the xaxis object and then change the tick labels
ax.xaxis.set_ticklabels(['M', 'F'], fontsize=20)
```

## Matrix plots
The same relationship exists between **`heatmap`**(axes plot) and its more powerful grid plot, **`clustermap`**. Matrix plots do not accept tidy data directly. The data needs to be pivoted or aggregated (like the data below).

```python
tips_pivot = tips.pivot_table(index='sex', columns='day', values='tip', aggfunc='mean')
tips_pivot
```

```python
sns.heatmap(tips_pivot, annot=True)
```

## clustermap
**`clustermap`** creates a heat map and performs hierarchical clustering at the same time. It will rearrange the column order so that the closest clusters of columns will appear together. The clustering is visualized with a dendogram outside of both the x and y axes. 

The **`college`** dataset below shows a strong cluster for math and verbal SAT scores. The lower the height of the clustering branch in the dendogram, the stronger the relationship. Pell grants and percent loans are the next group clustered together and are both negatively correlated with SAT scores.

```python
college = pd.read_csv('data/college.csv')
```

```python
sns.clustermap(college.corr())
# cg.ax_heatmap.set_yticklabels(cg.ax_heatmap.get_yticklabels(), rotation=0)
# plt.setp(cg.ax_heatmap.yaxis.get_majorticklabels(), rotation=0); # roatate the tick labels so they are readable
```

### Modifying a Seaborn grid
seaborn grids are a bit more difficult to modify than the seaborn **`axes`** plotting functions. The grid plotting functions return a seaborn grid object (it has Grid in its name) as opposed to a matplotlib axes directly. You must first access the individual plots inside a grid and then make your changes.

### Rotating the y-tick labels
The labels for the y-ticks are vertical and unreadable. To rotate them we must first latch onto the heatmap plot and then grab the labels and rotate them.

```python
# assign the result of the seaborn function to a variable.
cg = sns.clustermap(college.corr())

# seaborn grids usually have many matplotlib axes. Seaborn gives us directly access to the heatmp with ax_heatmap
ax = cg.ax_heatmap

# You have to grab the labels and then send those same labels to the set_ticklabels method
labels = ax.yaxis.get_ticklabels()
ax.yaxis.set_ticklabels(labels, rotation=0);
```

### Changing the size of seaborn grids
The grid seaborn plotting functions, have the **`size`** and **`aspect`** parameters that control the size. **`aspect`** is only needed if the width of each plot needs to be changed (and it will change to size * aspect).

```python
# default grid swarm plot
sns.factorplot(x='sex', y='tip', hue='smoker', col='day', data=tips, kind='swarm')
```

```python
# use the size parameter. its defaulted to 4
# it looks smaller but thats how its rendered in the notebook.
# if you save the plot with the savefig method you will see that it is infact larger

sns.factorplot(x='sex', y='tip', hue='smoker', col='day', data=tips, size=6, kind='swarm')
```

```python
# use the aspect parameter to change the width to be aspect * size
sns.factorplot(x='sex', y='tip', hue='smoker', col='day', data=tips, size=6, aspect=2, kind='swarm')
```

# pandas is sometimes the right choice for plotting
pandas DataFrame and Series objects provide the **`plot`** method for easy plotting. pandas, like seaborn uses matplotlib to make sensible plots. The pandas **`plot`** method is not as powerful nor as aesthetically appealing as seaborn. There are a few plots that pandas can do better than seaborn. 

The author of seaborn purposefully did not try and replicate all the functionality available in the pandas plot method. See [pandas documentation for plotting](http://pandas.pydata.org/pandas-docs/stable/visualization.html#subplots) for more.

Here are some things that pandas plot method does better than seaborn

* Ordered count plots
* Horizontal bar charts
* Wide data - one plot per column
* Stacked bar charts
* Time series line plots
* Time series scatter plots

pandas does not allow extra dimensional plotting and so does not have the seaborn equivalents for **`hue`**,  **`col`** or **`row`** parameters. Those parameters are part of the reason seaborn is so powerful. pandas also works much better with wide (aggregated or pivoted and not tidy) data than seaborn.

### Examples where pandas > seaborn

### Ordered and horizontal count plots
Seaborn **`countplot`** easily plots the counts of each level of any categorical variable. But there is some difficulty when wanting to order the bars.

```python
sns.countplot(x='day', data=tips)
```

We must manually set the order with the **`order`** parameter which forces us to compute the order from the Series **`value_counts`** method.

```python
order = tips.day.value_counts().index
sns.countplot(x='day', data=tips, order=order)
```

```python
# make it horizontal by using y=
order = tips.day.value_counts().index
sns.countplot(y='day', data=tips, order=order)
```

### Same but easier with pandas
Maybe that isn't so bad but pandas is slightly easier and capable in a single line. pandas defaults all plots to line plots. You specify the plot you would like with the **`kind`** paramter. You can directly add matplotlib arguments without actually invoking matplotlib directly.

Notice that pandas requires the aggregated total from **`value_counts`** and not the tidy DataFrame to plot the counts.

```python
# use rot to rotate the x labels
tips.day.value_counts().plot(kind='bar', rot=0, figsize=(8,6))
```

```python
# use barh for a horizontal
tips.day.value_counts().plot(kind='barh')
```

## Wide - aggregated or pivoted data
pandas works nicely with data that has been pivoted or aggregated from tidy data. pandas plots a new bar/line for every column.

```python
g = tips.groupby(['day', 'sex'])['tip'].mean().unstack()
g
```

```python
g.plot(kind='bar')
```

### Duplicating in seaborn
This is actually easier in seaborn if you have tidy data. But if the data has already been aggregated then this plot will not be easily done with seaborn.

```python
sns.barplot(x='day', y='tip', data=tips, hue='sex')
```

## Stacked bar 
The pandas plot method contains a **`stacked`** parameter. Setting this to **`True`** will stack the bars.

```python
# by default pandas will make a bar for every column. The index will be the x-axis
g.plot(kind='bar', stacked=True, rot=0)
```

```python
# move legend location. pandas plots return the same matplotlib axes object.
# Use it to modify the plot

ax = g.plot(kind='bar', stacked=True, rot=0)
ax.legend(loc='upper left')
```

```python
# no stacked barplot directly in seaborn
sns.countplot(x='day', data=tips, hue='sex')
```

```python
# can sort of reproduce this with seaborn by overlaying the plots but is very clumsy

sns.barplot(x='day', y='tip', data=tips, estimator=np.size, color='green')
sns.barplot(x='day', y='tip', data=tips.query('sex=="Male"'), estimator=np.size, color='blue')
```

# Time series and line plots with matplotlib and pandas
seaborn does in-fact have **`tsplot`** for time series but this only works for evenly spaced time series. It also does not format dates correctly for the tick labels and if you have unevenly distributed time series data **`tsplot`** will not work.

It turns out that the best plotting method for producing a scatter plot with time-series data is found directly in matplotlib. A function called **`plot_date`** exists. It is also doable with pandas **`plot`** with a workaround.

```python
# create some random time series data
n = 100
date = np.random.choice(pd.date_range('1-1-2014', '12-31-2016'), n)
num = np.random.rand(n)
df = pd.DataFrame(data={'date':date, 'num':num})
df.head(10)
```

### Using `plt.plot_date` to make a time-series scatter plot
The **`plot_date`** has a similar interface to seaborn's plotting functions. You pass it strings to the **`x`** and **`y`** parameters and your DataFrame to the **`data`** parameter.

```python
plt.xticks(rotation=45)
plt.plot_date(x='date', y='num', data=df)
```

### Use pandas to make a time-series scatter plot
Normally, you would set the **`kind`** parameter to **`scatter`** but pandas only accepts numeric and not date columns for scatter plots.

There is a small workaround that involves changing the default line style. Remember, that by default pandas plots a line. You can change the line style to **`o`** and it will look like a scatter plot.

```python
# default line plot
df.plot(x='date', y='num')
```

```python
# use the style parameter to make it appear to be a scatter plot
df.plot(x='date', y='num', style='o', ms=3)
```

### Use a pandas Series with a `DatetimeIndex` to plot time-series
If you have a pandas Series with a DatetimeIndex, you can immediately plot the time-series with the **`plot`** method.

```python
df.set_index('date').plot()
```

### pandas really excels with time-series line plots
If the DataFrame contains has a DatetimeIndex like the plot above it is easy to plot multiple time series. The DataFrame below has stock data from 3 different companeis. Each column gets its own line.

```python
import pandas_datareader as pdr
```

```python
aapl = pdr.DataReader('aapl', 'google')['Close'].rename('AAPL')
goog = pdr.DataReader('goog', 'google')['Close'].rename('GOOG')
slb = pdr.DataReader('slb', 'google')['Close'].rename('SLB')
df = pd.concat([aapl, goog, slb], axis=1)
```

```python
df.head()
```

```python
df.plot(figsize=(8,6))
```
