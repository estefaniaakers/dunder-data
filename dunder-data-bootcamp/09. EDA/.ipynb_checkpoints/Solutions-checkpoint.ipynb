# 0. Seaborn

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

college = pd.read_csv('data/college.csv')
tips = sns.load_dataset('tips')
```

### Problem 1
<span  style="color:green; font-size:16px">Using `college.csv`, plot the number of schools per state. Change the size of the figure so it looks good. Use the **`plt.xticks`** function to rotate the tick marks.</span>

```python
plt.xticks(rotation=90)
ax = sns.countplot(college.STABBR)
ax.figure.set_size_inches(10,6)
```

### Problem 2
<span  style="color:green; font-size:16px">Use pandas to create the same plot from problem 1, but make it an ordered horizontal bar chart.</span>

```python
college.STABBR.value_counts(ascending=True).plot(kind='barh', figsize=(16,10))
```

### Problem 3
<span  style="color:green; font-size:16px">Filter out only colleges that have the top 8 most total undergraduate student population (UGDS) and assign it to **`college_8`**. Then make a boxplot by state of those college populations.</span>

```python
top_8 = college.groupby('STABBR')['UGDS'].sum().sort_values(ascending=False).iloc[:8]
college_8 = college[college.STABBR.isin(top_8.index)]
college_8.head()
```

```python
sns.boxplot(x='STABBR', y='UGDS', data=college_8)
```

### Problem 4
<span  style="color:green; font-size:16px">The resulting plot from problem 3 looks terrible. Why does it look so bad? Transform the data in some way to make it look more reasonable.</span>

```python
# the outlier for AZ is skewing the data
# use a log scale for the y-axis or remove the outlier

# use the y-scale for the axis
ax = sns.boxplot(x='STABBR', y='UGDS', data=college_8)
ax.set_yscale('log')
```

```python
# removing the outlier does not do as well as taking the log
sns.boxplot(x='STABBR', y='UGDS', data=college_8.query('UGDS < 100000'))
```

### Problem 5
<span  style="color:green; font-size:16px">Using `college_8`, plot the linear relationship between verbal and math SAT scores using a 99.999% confidence interval. Make the figure larger than the default.</span>

```python
plt.figure(figsize=(10,8))
sns.regplot(x='SATVRMID', y='SATMTMID', data=college_8, ci=99.999)
```

### Problem 6
<span  style="color:green; font-size:16px">Using `college_8`, plot the linear relationship between verbal and math SAT scores. Plot a different colored regression line by State. Use [this stackoverflow post](http://stackoverflow.com/questions/25212986/how-to-set-some-xlim-and-ylim-in-seaborn-lmplot-facetgrid) to change the limits of the y-axis.</span>

```python
g = sns.lmplot(x='SATVRMID', y='SATMTMID', hue='STABBR', data=college_8)
g.set(ylim=(300, 800))
```

### Problem 7
<span  style="color:green; font-size:16px">Using `college_8`, plot the linear relationship between verbal and math SAT scores. Plot each state in its own panel and make two colored regression lines based on religious affiliation (RELAFFIL). Use the **`col_wrap`** parameter to fit all the panels easier. Do religiously affiliated schools do better or worse on their SATs?</span>

### Problem 8
<span  style="color:green; font-size:16px">Using the tips seaborn dataset, use a point plot to determine which which group, males or females tip more.</span>

```python
sns.pointplot(x='sex', y='tip', data=tips)
```

### Problem 9
<span  style="color:green; font-size:16px">Do the same point plot as problem 8, but add a new line for each type of smoker. Also, remove the confidence interval bars.</span>

```python
sns.pointplot(x='sex', y='tip', hue='smoker', data=tips, ci=0)
```

### Problem 10
<span  style="color:green; font-size:16px">Find all the combinations of sex, smoker, day and time where females tip more than males on average using only seaborn.</span>

There are 4 combinations where females tip more than males
* Friday lunch smoker
* Friday dinner non-smoker
* Saturday dinner smoker (about equal)
* Sunday Dinner non-smoker

```python
# Friday lunch smoker, Friday dinner non-smoker, Saturday dinner smoker (about equal), Sunday Dinner non-smoker
sns.factorplot(x='sex', y='tip', hue='smoker', data=tips, ci=0, kind='point', col='day', row='time')
```

### Problem 11
<span  style="color:green; font-size:16px">Find the solution to problem 10 using pandas without any plots</span>

```python
# Saturday dinner no was just a little higher for males
tips_grouped = tips.groupby(['day', 'time', 'smoker', 'sex'])['tip'].mean().unstack()
tips_grouped.round(2).style.highlight_max(axis=1)
```

### Problem 12
<span  style="color:green; font-size:16px">Comparing raw tip amounts is not ideal. Think of a different metric and redo the plot and the table from problems 10 and 11. Are the results the same?</span>

```python
# females tip higher as a percentage
tips['tip_percentage'] = tips.tip.div(tips.total_bill).round(3)
sns.factorplot(x='sex', y='tip_percentage', hue='smoker', data=tips, ci=0, kind='point', col='day', row='time')
```

```python
tips_grouped2 = tips.groupby(['day', 'time', 'smoker', 'sex'])['tip_percentage'].mean().unstack()
tips_grouped2.round(2).style.highlight_max(axis=1)
```
