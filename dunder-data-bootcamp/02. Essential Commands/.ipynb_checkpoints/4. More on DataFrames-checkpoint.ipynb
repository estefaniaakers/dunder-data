### In-Place operations
Many Series and DataFrame methods have the **`inplace`** parameter. It always defaults to **`False`**. When **`inplace`** is False, the operation returns some object (usually a Series or DataFrame - the same type of object that called the method).

When **`inplace`** is switched to **`True`**, the object **`None`** is returned and the changes happen in-place. The object calling the method is mutated. There is no need to use an assignment operation (the equals sign) when  **`inplace`** is True.

Let's first recall how the list method **`append`** modifies a list in-place.

```python
# modify a list in place
a = [1, 2, 3]

# append an element. No equal sign needed
a.append(4)
```

```python
# output the list
a
```

All the previous operations executed on a Series or a DataFrame have had **`inplace`** defaulted to **`False`**. Let's see how renaming a column in a DataFrame works when **`inplace`** is changed to True. 

```python
college = pd.read_csv('data/college.csv', index_col='INSTNM')
```

```python
# old command is: college = college.rename(columns={'UGDS':'UNDERGRAD_POP'})
# new command has inplace=True and no assignment to new variable
college.rename(columns={'UGDS':'UNDERGRAD_POP'}, inplace=True)
college.head()
```

### Warning about using inplace
Switching operations to **`inplace=True`** is atypical behavior and generally not preferred. Reassigning the output to a new (or the same) variable is preferable since this is the default behavior and is less confusing.

### Selecting columns by substring
It would be quite annoying to write out all 9 UGDS columns by hand. There are a few ways to programmatically select these columns with the easiest done with the **`filter`** method. Pass **`filter`** the string that you are searching for as the **`like`** argument. If you are familiar with regular expressions you can do more complex searches with the **`regex`** argument.

```python
# Get the top 5 of a series
college_ugds['UGDS_ASIAN'].nlargest(5)
```

### Missing Values
Missing values are represented by the NumPy object **`np.nan`** and are ignored in many aggregation methods like **`mean, median, sum`**,  etc... The **`count`** method does not include missing values either.

### How to count missing values
There are multiple ways of doing this. The **`isnull`** method converts all values to **`True/False`**. Summing the resulting DataFrame counts the missing values. **`notnull`** does the opposite.

```python
# change all values to True/False
college_ugds.isnull().head()
```

```python
# There are no missing values in the top 5 rows but there are at the tail
college_ugds.isnull().tail()
```

```python
# now just sum up every column
college_ugds.isnull().sum()
```

```python
# We can sum again to get the grand total of missing values
college_ugds.isnull().sum().sum()
```

```python
# add missing values to non-missing to get total number of rows in dataset
college_ugds.isnull().sum() + college_ugds.count()
```

## Operations across the rows
All the operations thus far have been calculated down a column. It is possible to perform most of the operations across the rows as well. A very important argument, **`axis`** controls which direction the operation happens. 

The **`axis`** argument can be  set to 0 or 1. When **`axis`** is 0, the operation will happen down the columns and when **`axis`** is 1 the operation will go across the rows.

If you think 0 and 1 are cryptic, you can use strings **index** or **columns**, respectively in their place. Pandas defaults these operations to work down the columns.

```python
# same behavior as default
college_ugds.sum(axis='index')
```

```python
# same behavior as default. Can use 0 instead of 'index'
college_ugds.sum(axis=0)
```

### New behavior - add across the rows

```python
# add the race percentages up. Should add up close to one
college_ugds.sum(axis='columns').head(15)
```

```python
# same as using axis=1
college_ugds.sum(axis=1).head(15)
```

### Changing Data Types
The **`college`** DataFrame has the column **`MD_EARN_WNE_P10`**, which from the data dicitonary is the median earnings after 10 years of enrollment. Looking at the top of this notebook when **`.info`** was executed, the data type came back as object.

But from its description and inspection it seems like it should be numeric. Series have a **`dtype`** attribute which returns the datatype

```python
# output data type of MD_EARN_WNE_P10
# use dtype for Series and dtypes for DataFrames

# 'O' means object
college['MD_EARN_WNE_P10'].dtype
```

```python
# lets inspect some of the values
college['MD_EARN_WNE_P10'].head(10)
```

### Sorting to see strings
Column **`MD_EARN_WNE_P10`** certainly appears to be numeric. Since numbers come first lexicographically, sorting the column by descending order would put any strings at the top.

```python
# so this is why this column was an object and not numeric
college['MD_EARN_WNE_P10'].sort_values(ascending=False).head()
```

### Converting to numeric
In the previous notebook, the method **`astype`** was used to convert a string to a datetime. **`astype`** is not a robust method and will error when the conversion can't happen. Pandas provides helper functions that give more flexibility when converting types. 

The **`to_numeric, to_datetime`** and **`to_timedelta`** functions are much more robust and give much more control to handle different situations where **`astype`** would error out.

The **`to_numeric`** parameter **`errors`** can be set to 'coerce' to yield a missing value every time it encounters a value it cannot convert such as the string 'PrivacySuppressed'.

```python
# convert column to numeric. make strings missing values
college['MD_EARN_WNE_P10'] = pd.to_numeric(college['MD_EARN_WNE_P10'], errors='coerce')

# use describe to check type and see summary stats
college['MD_EARN_WNE_P10'].describe()
```

## Impress your friends with DataFrame styling
[Styling the appearance of the DataFrame](http://pandas.pydata.org/pandas-docs/stable/style.html) output is a new feature and can really bring attention to particular parts of the DataFrame.

One clear use-case is to highlight the max/min values of numeric columns. Let's select only the numeric columns with the **`select_dtypes`** method.

```python
college_numeric = college.select_dtypes(include=['number'])
college_numeric.head()
```

### Dropping Binary Columns
Several columns have only 0/1 values. These are not interesting for taking the max/min. Lets find all the columns that have at most two unique values.

```python
is_binary = college_numeric.nunique() == 2
is_binary.head()
```

### Use boolean indexing on the itself
Place the **`is_binary`** Series inside its own indexing operator to filter for the binary columns. Then select the index and use the **`tolist`** index method to put the binary columns in a list.

```python
binary_cols = is_binary[is_binary].index.tolist()
binary_cols
```

### Use the drop method
Now we can use the drop method to remove these columns

```python
college_numeric_drop = college_numeric.drop(binary_cols, axis='columns')
college_numeric_drop.head()
```

### Use idxmax to find the college with the maximum value for each column
The **`idxmax`** method to return th

```python
# get index label for max value of every row
idx_max = college_numeric_drop.idxmax()
idx_max
```

### Only need unique values
Since we don't want to duplicate any rows we can grab the unique values.

```python
max_cols = idx_max.unique()
max_cols
```

### Select only schools with maximum with .loc

```python
# use .loc to select only these rows
college_numeric_drop.loc[max_cols].head()
```

### Use the style attribute
The DataFrame style attribute opens up access to many methods that style DataFrames. The **`style.highlight_max()`** method will make it very easy to see where the max value for each column occurs.

```python
# impress friends time
college_numeric_drop.loc[max_cols].style.highlight_max()
```

### Filling Missing Values
Pandas gives great flexibility for replacing missing values with some other value. The **`fillna`** method works for both DataFrames and Series. To clearly show the versatility of **`fillna`** a small DataFrame will be created and operated upon.

```python
df_test = pd.DataFrame({'City':['Houston', np.nan, np.nan, 'New Orleans', 'Los Angeles'],
                       'State':['Texas', 'Maine', np.nan, 'Louisiana', 'California'],
                       'Pop':[100, 50, 30, np.nan, 80]},
                      columns=['City', 'State', 'Pop'])

df_test
```

```python
# count the missing values
df_test.isnull().sum()
```

```python
# fillna with a constant
# non-sensical with both string and numeric columns but still works
df_test.fillna(0)
```

```python
# fill with constant string
# notice how data type of Pop is now object

df_test.fillna('Nevada')
```

```python
# use the `method` argument to fill using previous/next values

# Takes closest missing value in front for each column
df_test.fillna(method='backfill')
```

```python
# go the other way
df_test.fillna(method='ffill')
```

```python
# A common approach to missing values is to replace them by the column mean
# This only works for numeric columns

df_test.fillna(df_test.mean())
```

```python
# can also use for Series
# and limit how many missing values get filled

df_test['City'].fillna('Bangor', limit=1)
```
